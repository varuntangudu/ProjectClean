```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd

game_stats = pd.read_csv("../data/processed/game_stats.csv")
```

```python
# Distribution of Total Yards Gained per Game
print(game_stats.columns)  # Add this line to check available columns
sns.histplot(game_stats['total_yards'], kde=True)
plt.title("Distribution of Total Yards Gained per Game")
plt.xlabel("Total Yards Gained")
plt.ylabel("Frequency")
plt.show()
```

```python
# Yards Gained vs. Win Outcome
sns.barplot(data=game_stats, x='win', y='total_yards')
plt.title("Yards Gained by Win Outcome")
plt.xlabel("Win (1) / Loss (0)")
plt.ylabel("Total Yards Gained")
plt.show()
```

```python
# Feature Correlation Heatmap
corr = game_stats.corr(numeric_only=True)
plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()
```
